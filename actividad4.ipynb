3. Desarrollo del Programa:
Implementar el programa en Python para gestionar tareas utilizando solo listas y diccionarios.

• Agregar al menos tres tareas distintas al programa utilizando el método insert() de la lista. • Cada tarea debe ser representada como un diccionario con las claves nombre, descripcion y estado.


```python
tareas = []

tareas.insert(0, {"Nombre": "Tarea 1", "Desripcion": "Leer Clase 4","Estado": "Realizado"})
tareas.insert(1, {"Nombre": "Tarea 2", "Desripcion": "Practicar ramas","Estado": "Realizado"})
tareas.insert(2, {"Nombre": "Tarea 3", "Desripcion": "Hacer act 4","Estado": "En proceso"})
tareas.insert(3, {"Nombre": "Tarea 4", "Desripcion": "Subir act","Estado": "Pendiente"})

print("Tareas agregadas:")
for tarea in tareas:
    print("Nombre:", tarea["Nombre"], "- Estado:", tarea["Estado"])
```

Tareas agregadas:
Nombre: Tarea 1 - Estado: Realizado
Nombre: Tarea 2 - Estado: Realizado
Nombre: Tarea 3 - Estado: En proceso
Nombre: Tarea 4 - Estado: Pendiente
• Eliminación de Tareas: Utilizar el método pop() de la lista para eliminar una de las tareas ingresadas.

```python
tarea_eliminada = tareas.pop(1)

print("Se eleimino la tarea: ", tarea_eliminada)
```

Se eleimino la tarea:  {'Nombre': 'Tarea 2', 'Desripcion': 'Practicar ramas', 'Estado': 'Realizado'}
• Mostrar las Tareas: Imprimir en pantalla las tareas restantes almacenadas en la lista, accediendo directamente a los elementos mediante sus índices

```python
print("Las tareas restantes son: ")
for i in range(len(tareas)):
    print(f"Tarea {i+1}: {tareas[i]}")
```

Las tareas restantes son: 
Tarea 1: {'Nombre': 'Tarea 1', 'Desripcion': 'Leer Clase 4', 'Estado': 'Realizado'}
Tarea 2: {'Nombre': 'Tarea 3', 'Desripcion': 'Hacer act 4', 'Estado': 'En proceso'}
Tarea 3: {'Nombre': 'Tarea 4', 'Desripcion': 'Subir act', 'Estado': 'Pendiente'}
